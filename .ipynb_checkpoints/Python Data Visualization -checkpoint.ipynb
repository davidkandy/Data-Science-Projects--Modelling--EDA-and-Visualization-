```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np

x = np.linspace(0,10,25)
y = x * x + 2
print(x)
print(y)
print(np.array([x,y]).reshape(25, 2))
```

```python
pylab.plot(x,y, 'r')
```

```python
"""Drawing a sub-graph"""
pylab.subplot(1, 2, 1) #The contents of the bracket represents (row, column, index)
pylab.plot(x, y, 'r--') #The third parameter here determines color and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x,'g*-')
```

```python
pylab.subplot(3, 2, 1) #The contents of the bracket represents (row, column, index)
pylab.plot(x, y, 'r--') #The third parameter here determines color and line style
pylab.subplot(1, 2, 1)
pylab.plot(y, x,'g*-')
pylab.subplot(3, 2, 2)
pylab.plot(y, x,'b*-')
```
