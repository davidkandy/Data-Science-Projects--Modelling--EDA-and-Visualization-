```python
''' Draw a Histogram'''


n = np.random.randn(100000)
fig, axes = plt.subplots(1,2,figsize=(12, 4))

axes[0].set_title("Default Histogram")
axes[0].hist(n, width=.79)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=8)
```

```python
# Draw a Contour map

import matplotlib
import numpy as np
import matplotlib.cm as cm # cm = color map
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta) # arange() creates an array of numbers from -3.0 to 3.0 with an increment of delta's value i.e 0.025
y = np.arange(-2.0, 2.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

print(X)
print("\n")
print(Y)
```

```python
fig, ax = plt.subplots()

CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title("Contour Map")
```

```python
# 3D Surface image

from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14, 6))

# Specify the 3D graphics to draw, with projection='3d'
ax = fig.add_subplot(1, 2, 1, projection = '3d') # add_subplot(column, width, index)
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
```

# Write a Python program to create a pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt

# Data to plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'brown']
```

```python
explode = (0.1, 0, 0, 0, 0, 0)

# plt.pie(popularity)

plt.pie(popularity, explode=explode, labels=languages, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
```
